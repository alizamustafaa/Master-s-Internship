```python
import mtneedlet as nd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
B= 2
nside=512
lmax = 2*nside
j_min=0
j_max = 10
b_max = 5
n_bands = j_max - b_max + 1
b_in = nd.standardneedlet(B,list(range(j_min,j_max+1)),lmax)
# b = nd.mexicanneedlet(B,list(range(j_min,j_max+1)),lmax)

b=np.zeros((n_bands,lmax+1))
b[0,:]=np.sqrt(np.sum((b_in**2)[:b_max+1,:lmax+1],axis=0))
for i in range(1,n_bands):
    b[i]=b_in[b_max+i]
    
b[0,:2]=1.
```

```python
nd.plot_bl(b)
nd.plot_blprofile
plt.xscale('log')
plt.xlim(2,lmax)
```
